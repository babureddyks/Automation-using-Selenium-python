```python
import time 
import string
import openpyxl
import os
from selenium.webdriver.common.keys import Keys
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support.expected_conditions import presence_of_element_located
from selenium.common.exceptions import NoSuchElementException
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.chrome.options import Options


results = []
searchfor=['Nepali stores', 'Indian stores', 'South Asian stores', 'Pakistani stores', 'Bangladeshi stores', 'Sri Lankan stores']
cities=small

options = webdriver.ChromeOptions()

opt = Options()
opt.add_argument("--disable-infobars")
opt.add_argument("--disable-extensions")
options.add_experimental_option("prefs", { \
    #"profile.default_content_setting_values.media_stream_mic": 2, 
    #"profile.default_content_setting_values.media_stream_camera": 2,
    "profile.default_content_setting_values.geolocation": 2, 
    #"profile.default_content_setting_values.notifications": 2 
  })

options.add_argument('--deny-permission-prompts')
driver = webdriver.Chrome(ChromeDriverManager().install(),chrome_options=options)
driver.get('https://www.google.com/maps/@37.6,-95.665,4z')
time.sleep(5)





def page():
   #smooth scrolling
    for i in [11,21,25,31,35,41]:
        try:
            a=driver.find_element_by_xpath("//*[@id='pane']/div/div[1]/div/div/div[2]/div[1]/div["+str(i)+"]/div/a")
            driver.execute_script("arguments[0].scrollIntoView();",a)
            time.sleep(5)
        except NoSuchElementException:
            pass

    
    page_content = driver.page_source
    from parsel import Selector
    response = Selector(page_content)

    #results = []

    for i in range(3,61,2):
    #print(i)
    
        for el in response.xpath('//*[@id="pane"]/div/div[1]/div/div/div[2]/div[1]/div['+str(i)+']'):
            results.append({
                'Title': el.xpath('//*[@id="pane"]/div/div[1]/div/div/div[2]/div[1]/div['+str(i)+']/div/div[2]/div[2]/div[1]/div/div/div/div[1]/div/span/text()').extract_first(),
                'Address': el.xpath('//*[@id="pane"]/div/div[1]/div/div/div[2]/div[1]/div['+str(i)+']/div/div[2]/div[2]/div[1]/div/div/div/div[4]/div[1]/span[2]/jsl/span[2]/text()').extract_first(),
                'Phone': el.xpath('//*[@id="pane"]/div/div[1]/div/div/div[2]/div[1]/div['+str(i)+']/div/div[2]/div[2]/div[1]/div/div/div/div[4]/div[2]/span[3]/jsl/span[2]/text()').extract_first(''),
                'Phone2': el.xpath('//*[@id="pane"]/div/div[1]/div/div/div[2]/div[1]/div['+str(i)+']/div/div[2]/div[2]/div[1]/div/div/div/div[4]/div[2]/span[2]/jsl/span[2]/text()').extract_first(''),
                'Category': el.xpath('//*[@id="pane"]/div/div[1]/div/div/div[2]/div[1]/div['+str(i)+']/div/div[2]/div[2]/div[1]/div/div/div/div[4]/div[1]/span[1]/jsl/span[2]/text()').extract_first(),
                'City':ii
                
                    })
            #time.sleep(5)
            
    #next button
    
    try:
        abc=driver.current_url[:70]
        time.sleep(5)
        elem =driver.find_element_by_xpath('/html/body/div[3]/div[9]/div[8]/div/div[1]/div/div/div[2]/div[2]/div/div[1]/div/button[2]') 
        if elem.is_enabled():
            driver.execute_script("arguments[0].click();", elem)

            #elem.click()
            time.sleep(5)
            if abc==driver.current_url[:70]:
                page()
                #time.sleep(5)
    except NoSuchElementException:
            pass
            
    
    return
#end page loop
for ii in cities:

    for m in searchfor:
        driver.get('https://www.google.com/maps/@37.6,-95.665,4z')
        time.sleep(5)
        #driver.find_element_by_id('searchboxinput').clear()
        searchbox=driver.find_element_by_id('searchboxinput')
        location= ii
        searchbox.send_keys(location)
        time.sleep(2)
        
        searchbox.send_keys(Keys.ENTER)
        time.sleep(6)
        
        Place = driver.find_element_by_class_name("tactile-searchbox-input")
        driver.find_element_by_class_name("tactile-searchbox-input").clear()
        Place.send_keys(m)
        Submit = driver.find_element_by_xpath("//*[@id='searchbox-searchbutton']")
        #Submit.click()
        WebDriverWait(driver, 20).until(EC.element_to_be_clickable((By.XPATH, "//*[@id='searchbox-searchbutton']"))).click()
        Submit.click()
        time.sleep(3)
        
        page()
        time.sleep(3)
        
#drivers()
#time.sleep(6)


import pandas as pd
df=pd.DataFrame(results)        
df.to_excel("finaloutput.xlsx")


```
